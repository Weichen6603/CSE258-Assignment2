```python
import pandas as pd

# Function to filter the dataset by userId range and save to a new CSV file
def filter_and_save_ratings(input_csv, output_csv, x):
    """
    Filters the ratings dataset to include only rows with userId in range [1, x],
    and saves the result to a new CSV file.

    Parameters:
    - input_csv (str): Path to the input CSV file.
    - output_csv (str): Path to save the filtered CSV file.
    - x (int): Maximum userId to include.
    """
    # Load the dataset
    ratings = pd.read_csv(input_csv)
    
    # Filter by userId
    filtered_ratings = ratings[ratings['userId'] <= x]
    
    # Save the filtered dataset to a new file
    filtered_ratings.to_csv(output_csv, index=False)
    return filtered_ratings
```

```python
input_csv_path = "./dataset/ratings.csv"
output_csv_path = "./dataset/ratings_mid.csv"
filter_and_save_ratings(input_csv_path, output_csv_path, x=6000)
```
